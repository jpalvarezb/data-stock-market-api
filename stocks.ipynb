```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import requests
url = 'https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/day/2024-10-29/2025-01-27'
headers = {
    'Authorization':'Bearer FBSi0iXQoTK6F6kdDzyvAVOPRH7u_ZKq'
}
data = requests.get(url,headers=headers).json()
apple_df = pd.DataFrame(data['results'])
apple_df['t'] = pd.to_datetime(apple_df['t'], unit ='ms')
apple_df.set_index(apple_df['t'], drop = True, inplace=True)
apple_df
```

```python
apple_df = apple_df.rename(columns={ 'o':'open',
                         'c': 'close',
                         'h': 'high',
                         'l': 'low',
                         'n': 'number',
                         'v': 'volume',
                         'vw': 'avg_price'})
apple_df[['open', 'close', 'high','low']].plot(kind='bar',figsize=(15,5)).set_ylabel('Stock Price (USD)')
```
